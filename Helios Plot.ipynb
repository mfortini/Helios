```python
import pandas as pd
import plotly.express as px
import plotly.io as pio
```

```python
d=pd.read_csv('db.stat',header=None)
d.columns=['timestamp','signal','noise_nb','noise_wb']
d['data']=pd.to_datetime(d['timestamp'])
d.index=d['data']
d.drop(columns=['timestamp'],inplace=True)
```

```python
fig=px.line(d,y="signal");
fig.update_layout(template='plotly_white');
fig.update_layout(title='Letture sensore Helios Osservatorio Astronomico di Cento. Astrofili Centesi');
```

```python
pio.write_html(fig, file='index.html', auto_open=False);
```

```python
fig.show()
```
